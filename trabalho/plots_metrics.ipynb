```python
import matplotlib.pyplot as plt
import numpy as np

# Métricas para cada modelo
models = ['SVM', 'KNN', 'NC', 'XGBoost', 'Random Forest']
accuracy = [0.92, 0.87, 0.79, 0.99, 0.96]
precision_class_1 = [0.87, 0.80, 0.71, 0.98, 0.94]
recall_class_1 = [0.97, 0.96, 0.93, 0.99, 0.97]
f1_score_class_1 = [0.92, 0.87, 0.81, 0.99, 0.96]
precision_class_0 = [0.97, 0.96, 0.91, 1.00, 0.97]
recall_class_0 = [0.87, 0.79, 0.66, 0.98, 0.95]
f1_score_class_0 = [0.92, 0.86, 0.77, 0.99, 0.96]

# Criar gráfico de barras para a precisão
plt.figure(figsize=(10, 6))
plt.bar(models, precision_class_1, color='blue', alpha=0.7, label='Precision (Class 1)')
plt.title('Precision (Class 1) Comparison')
plt.xlabel('Model')
plt.ylabel('Precision')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para o recall
plt.figure(figsize=(10, 6))
plt.bar(models, recall_class_1, color='green', alpha=0.7, label='Recall (Class 1)')
plt.title('Recall (Class 1) Comparison')
plt.xlabel('Model')
plt.ylabel('Recall')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para o F1-score
plt.figure(figsize=(10, 6))
plt.bar(models, f1_score_class_1, color='orange', alpha=0.7, label='F1-score (Class 1)')
plt.title('F1-score (Class 1) Comparison')
plt.xlabel('Model')
plt.ylabel('F1-score')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para a precisão da classe 0
plt.figure(figsize=(10, 6))
plt.bar(models, precision_class_0, color='blue', alpha=0.7, label='Precision (Class 0)')
plt.title('Precision (Class 0) Comparison')
plt.xlabel('Model')
plt.ylabel('Precision')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para o recall da classe 0
plt.figure(figsize=(10, 6))
plt.bar(models, recall_class_0, color='green', alpha=0.7, label='Recall (Class 0)')
plt.title('Recall (Class 0) Comparison')
plt.xlabel('Model')
plt.ylabel('Recall')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para o F1-score da classe 0
plt.figure(figsize=(10, 6))
plt.bar(models, f1_score_class_0, color='orange', alpha=0.7, label='F1-score (Class 0)')
plt.title('F1-score (Class 0) Comparison')
plt.xlabel('Model')
plt.ylabel('F1-score')
plt.ylim(0, 1)
plt.legend()
plt.show()

# Criar gráfico de barras para a acurácia
plt.figure(figsize=(10, 6))
plt.bar(models, accuracy, color='red', alpha=0.7, label='Accuracy')
plt.title('Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()
plt.show()
```
